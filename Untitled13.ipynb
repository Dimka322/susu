```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd
import math
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
!unzip '/content/hiring.zip'
```

```python
data = pd.read_csv('/content/hiring.csv')
iris = pd.read_csv('/content/iris.csv')
```

```python
data = data.iloc[:, 1:]
```

```python
le = LabelEncoder()
```

```python
iris['SpeciesNum'] = le.fit_transform(iris['Species'])
```

```python
iris.SpeciesNum.value_counts()
```

```python
dataIris = iris[['Sepal-Length', 'Sepal-Width', 'Petal-Length', 'Petal-Width']]
labelIris = iris['Species']
```

```python
linkageTypes = ['single', 'complete', 'average', 'ward']
```

```python
for linkageType in linkageTypes:
  plt.figure(figsize=(12, 6))
  Dendrogram = dendrogram((linkage(data, linkageType)))
  plt.show()
```

```python
labels = list(set(labelIris))
labelNum = []
for i in labelIris:
  labelNum.append(labels.index(i))
```

```python
for linkageType in linkageTypes:
  plt.figure(figsize=(12, 6))
  Dendrogram = dendrogram((linkage(dataIris, linkageType)), labels=iris['SpeciesNum'].to_list(), leaf_font_size=7)
  plt.xticks(rotation=45)
  plt.show()
```
